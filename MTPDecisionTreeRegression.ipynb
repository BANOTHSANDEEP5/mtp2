```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
dataset = pd.read_excel('data.xlsx')
```

```python
dataset = dataset.iloc[:, 2:-1]
```

```python
plt.figure(figsize=[30, 30])

# Loop through columns
for i, col in enumerate(dataset.columns):
    plt.subplot(4, 3, i + 1)  # Use 'i' to index the subplots
    sb.barplot(data=dataset, x=col, y='Porosity (%)')  # Swap x and y for barplot
    plt.title(f'Porosity for {col}')  # Add a title for each subplot

plt.tight_layout()  # Adjust the layout for better spacing
plt.show()
```

```python
corrMatrix = dataset.corr()
plt.gcf().set_size_inches(10,9)
sb.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
target = dataset['Porosity (%)']
features = dataset.drop('Porosity (%)',axis=1)
```

```python
target
```

```python
features
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(features,target, test_size = 0.25)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_randomforest = RandomForestRegressor(random_state=0)
regressor_randomforest.fit(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor_decisiontree = DecisionTreeRegressor(random_state=0)
regressor_decisiontree.fit(X_train,y_train)
```

```python
y_predict_randomforest = regressor_randomforest.predict(X_test)
```

```python
y_predict_decisiontree = regressor_decisiontree.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score

mse = mean_squared_error(y_test, y_predict_randomforest)
rmse = np.sqrt(mse)
r2_score(y_test ,y_predict_randomforest)

print('Mean squared Error:', mse)
print('Root Mean squared Error:', rmse)
print('R-squared:', rmse)
```

```python
from sklearn.metrics import mean_squared_error , r2_score

mse = mean_squared_error(y_test, y_predict_decisiontree)
rmse = np.sqrt(mse)
r2_score(y_test ,y_predict_decisiontree)

print('Mean squared Error:', mse)
print('Root Mean squared Error:', rmse)
print('R-squared:', rmse)
```

```python
temp = [(y_test.values[i], y_predict_decisiontree[i]) for i in range(len(y_predict_decisiontree))]
temp = sorted(temp, key=lambda x: x[0])
temp1 = [u for (u,v) in temp]
temp2 = [v for (u,v) in temp]
x_axis = np.array(range(len(temp1)))

plt.plot(x_axis, temp1 , color = 'red', label='y_test')
plt.plot(x_axis, temp2 , color = 'blue', label='y_predict')

plt.title('y_test vs y_predict comparison')
plt.xlabel('total number of test cases')
plt.ylabel('y_test and y_predict')
plt.legend()
plt.show()
```
